```python
type1 = input()
type2 = input()

count = 0
ch = 0 

while ch < len(type1) and ch < len(type2): #while ch is less than the length of type 1 and type 2
     if type1[ch] == type2[ch]:     #if type1 index of ch is equal to the index of ch in type2 
         count += 1                #add 1 to the count
     ch +=1                        #and add 1 to the ch
if count > 0:    
    print (f"{count} character matches")
else:
    print ("0 characters match")
```
