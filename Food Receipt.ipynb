```python
food_name = input ("Enter food item name:\n")

food_price =  float (input("Enter item price:\n"))

food_quantity = int (input("Enter item quantity:\n")) 

total = food_price * food_quantity

print ("RECEIPT\n") 
print (f"{food_quantity} {food_name} @ ${food_price:.2f} = ${total:.2f}")

print (f"Total Cost: ${total:.2f}\n ")
 

food_name2 = input ("Enter second food item name:\n")

food_price2 = float (input ("Enter item price:\n"))

food_quantity2 = int (input ("Enter item quanttity\n"))

total2 = food_price2 * food_quantity2

print ("RECEIPT\n")

print (f"{food_quantity} {food_name} @ ${food_price:.2f} = ${total:.2f}\n")

print (f"{food_quantity2} {food_name2} @ ${food_price2:.2f} = ${total2:.2f}\n")

total_cost = total + total2

print (f"Total Cost: ${total_cost:.2f}")


                        
            
```
